```python
# import load_dotenv
from dotenv import load_dotenv

load_dotenv(override=True) # pass override=True to override existing variables and use the new ones from .env file
```

```python
import os 


#  check the github token and base url  for openai 

token = os.getenv("GITHUB_TOKEN")
base_url = os.getenv("GITHUB_BASE_URL")

if token and base_url:
    print(f"Github token and base url are set")
    print(f"Token: {token[:8]}")
    print(f"Base URL: {base_url}")
else:
    print("Github token or base url is not set check .env file")

    
```

```python
# now we will create openai client instance 
from openai import OpenAI

openai_client = OpenAI(base_url=base_url,api_key=token)
```

```python
# Create a list of messages in the familiar OpenAI format

messages = [{"role": "user", "content": "What is 2+2?"}]
```

```python
response = openai_client.chat.completions.create(
    model="gpt-4.1-nano",
    messages=messages
)

print(response.choices[0].message.content)
```

```python
# And now - let's ask for a question:

question = "Please propose a hard, challenging question to assess someone's IQ. Respond only with the question."
messages = [{"role": "user", "content": question}]
```

```python
# ask it - this uses GPT 4.1 mini, still cheap but more powerful than nano
response = openai_client.chat.completions.create(
    model="gpt-4.1-mini",
    messages=messages
)

question = response.choices[0].message.content

print(question)
```

```python
# First create the messages:

messages = [{"role": "user", "content": "pick a business area that might be worth exploring for an agentic ai opportunity"}]

#  the first call:
response = openai_client.chat.completions.create(
    model="gpt-4.1-mini",
    messages=messages
)

# Then read the business idea:

business_idea = response.choices[0].message.content

print(business_idea)
```

```python
# asking llm  another question on business idea 
question = " \n present a pain-point in that industry - something challenging that might be ripe for an Agentic solution."

messages = [{"role": "user", "content": business_idea+question}]

response = openai_client.chat.completions.create(
    model="gpt-4.1-mini",
    messages=messages
)

answer =  response.choices[0].message.content
print(answer)
```

```python
# now ask for solutin 
message = f'''Business Idea : {business_idea}  

{answer}

now propose the agentic solution 
'''

messages = [{"role": "user", "content":message}]

response = openai_client.chat.completions.create(
    model="gpt-4.1-mini",
    messages=messages
)


solution = response.choices[0].message.content
```
