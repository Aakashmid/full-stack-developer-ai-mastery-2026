The Agents SDK has a very small set of primitives:

- `Agents`, which are LLMs equipped with instructions and tools
- `Agents as tools / Handoffs`, which allow agents to delegate to other agents for specific tasks
- `Guardrails`, which enable validation of agent inputs and outputs

```python
# the imports 
from dotenv  import load_dotenv 
from agents import Agent , Runner , trace , set_default_openai_client , set_tracing_disabled , OpenAIChatCompletionsModel
from openai import OpenAI , AsyncOpenAI
import os
```

```python
load_dotenv(override=True)
```

```python

client = AsyncOpenAI(api_key=os.getenv("GITHUB_TOKEN"),base_url=os.getenv("GITHUB_BASE_URL"))


# set_default_openai_client(client)
set_tracing_disabled(disabled=True)   
```

```python
#  checking async client working or not 
response = await client.chat.completions.create(
    model="gpt-4o-mini",
    messages=[
        {"role":"user","content":"Hello world"}
    ]
    )
print(response.choices[0].message.content)
```

```python

model = OpenAIChatCompletionsModel(model="gpt-4o-mini",openai_client=client)  #

# Make an agent with name, instructions, model
agent = Agent(name="Jokester", instructions="You are a joke teller", model=model)
```

```python
result = await Runner.run(agent, "Tell a joke about Autonomous AI Agents")
result.final_output
```
