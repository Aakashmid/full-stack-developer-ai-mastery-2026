# Embeddings: Converting Text to Vectors

## What are Embeddings?

Embeddings are **numerical vector representations** of text that capture semantic meaning. They transform words, sentences, or documents into arrays of numbers that computers can mathematically compare.

**Key Idea:** Similar meanings → vectors positioned close together in vector space

---

## How Embeddings Work

### Example
 Text 1: "How to reset my password?"   
 Text 2: "I forgot my login password"    
 Text 3: "What is today's weather?"

 After Embedding:   
 Text 1 & 2 → Similar vectors (close proximity) → Related topics  
 Text 3 → Different vector (far away) → Unrelated


---

## Why Embeddings Matter

1. **Semantic Understanding** - Captures meaning, not just keywords
2. **Similarity Comparison** - Find related documents mathematically
3. **Efficiency** - Convert complex text into compact numbers
4. **ML Ready** - Numbers work with machine learning models

---

## Real-World Applications

### 1. **Semantic Search**
- User searches: "How do I change my password?"
- System finds: "Reset password tutorial" (similar meaning)
- Traditional keyword search would miss this

### 2. **RAG (Retrieval-Augmented Generation)**

- Step 1: Convert documents → Embeddings → Store in vector database
- Step 2: User asks question → Convert to embedding
- Step 3: Find similar document embeddings (closest vectors)
- Step 4: Send matched documents to LLM for answer generation


- Product embeddings + User query embedding → Find similar products
- Movie embeddings → Recommend similar movies

### 4. **Duplicate Detection**
- Compare embeddings of documents
- High similarity = Likely duplicate content

### 5. **Chatbot Understanding**
- Convert user messages → embeddings
- Match with training data embeddings
- Provide contextual responses

---

## Embedding Models Used Here

- **sentence-transformers/all-MiniLM-L6-v2** - Fast, lightweight (384 dimensions)
- **sentence-transformers/all-mpnet-base-v2** - Better quality (768 dimensions)

## Quick Workflow in This Notebook

1. Load HuggingFace API credentials
2. Initialize embedding model
3. Convert documents → vectors (embed_documents)
4. Convert queries → vectors (embed_query)
5. Compare similarity between vectors

```python
import os
from dotenv import load_dotenv, find_dotenv

_ = load_dotenv(find_dotenv())

# Get HuggingFace API key from environment
hf_api_key = os.getenv("HF_API_KEY")
```

```python
from langchain_huggingface import HuggingFaceEndpointEmbeddings

# Initialize HuggingFace Hub embeddings (with API key) ( to get api key go to hugging face and create an account )
embeddings = HuggingFaceEndpointEmbeddings(
    model="sentence-transformers/all-MiniLM-L6-v2",
    task="feature-extraction",
    huggingfacehub_api_token = hf_api_key,
)

# Test with documents
texts = [
    "How to reset my password?",
    "I forget my login password",
    "I do sketch sometimes"
]

vectors = embeddings.embed_documents(texts)  # embed_documents take a list of texts and return list of vectors

print(f"Number of text vectors: {len(vectors)}")
print(f"Embedding vector size: {len(vectors[0])}")

for vector in vectors:
    print(vector[:5])
```

```python
# Test query embedding
query = "Universe is vast and mysterious . In this vast universe , there are many galaxies and stars."
query_vector = embeddings.embed_query(query)  # embed_query takes a single text string and returns a single vector

print(f"First 5 values: {query_vector[:5]}")
```
