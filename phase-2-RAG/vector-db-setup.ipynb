```python
## Vector Database Setup with LangChain 
# ( using ChromaDb as vector db )
```

```python
import os
from dotenv import load_dotenv
from langchain_huggingface import HuggingFaceEndpointEmbeddings
from langchain_chroma import Chroma

load_dotenv()

hf_api_key = os.getenv("HF_API_KEY")
```

```python
# emmedding model from HuggingFace
embeddings = HuggingFaceEndpointEmbeddings(
    model="sentence-transformers/all-MiniLM-L6-v2",
    task="feature-extraction",
    huggingfacehub_api_token = hf_api_key,
)

# Chroam vector db setup ( using local persistence directory )
vectorstore = Chroma(
    collection_name="example_collection",
    embedding_function=embeddings,
    persist_directory="./local_chromaDb/"
)


# for cloud-based ChromaDb setup
# vectorstore = Chroma(
#     collection_name="example_collection",
#     embedding_function=embeddings,
#     client_settings= Settings(
#     chroma_api_impl="rest",
#     chroma_server_host="api.trychroma.com",
#     chroma_server_http_port=443
#     )
# )
```

```python
# add data
texts = [
    "Quantum mechanics explores the behavior of particles at atomic and subatomic scales.",
    "Python is a versatile programming language widely used in data science and web development.",
    "The Roman Empire was one of the most influential civilizations in world history.",
    "Shakespeare's plays often explore themes of love, betrayal, and power.",
    "Mount Everest, located in the Himalayas, is the tallest mountain in the world.",
    "Stoicism teaches resilience by focusing on what is within our control.",
    "Lionel Messi is considered one of the greatest football players of all time."
]

vectorstore.add_texts(texts)
```

```python
# query / similarity search 


query = "highest mountains"
# query = "football "
results = vectorstore.similarity_search(query, k=2)

for doc in results:
    print(doc.page_content)
```
